```python
# Importing required liabraries
import re
import pandas as pd
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Initialize Edge WebDriver & finding all the ads from the page
driver = webdriver.Edge()
url = 'https://www.google.com/search?q=samsung+mobile'
driver.get(url)
data = driver.find_elements(By.CLASS_NAME, 'mnr-c')
ads = [ad for ad in data if not ad.get_attribute('class').startswith('view-all')]
```

```python
# Storing the data into lists and then in the data frame

model_names = []
ram_datas = []
prices = []
ecom_sites = []
shipping_types = []
image_urls = []
product_links = []

# Loop through each ad again and extract the required information
for ad in ads:
    model_name = ad.find_element(By.CSS_SELECTOR, 'span.pymv4e').text
    price = ad.find_element(By.CSS_SELECTOR, 'div.T4OwTb span.e10twf').text
    ecom_site = ad.find_element(By.CSS_SELECTOR, 'div.LbUacb span.zPEcBd').text
    product_link = ad.find_element(By.CSS_SELECTOR, '.plantl').get_attribute('href')
    image_url = ad.find_element(By.CSS_SELECTOR, 'div.D6nsM img').get_attribute('src')
    
    model_name = ad.find_element(By.CSS_SELECTOR, 'span.pymv4e').text
    specs = model_name.splitlines()
    for ram in specs:
        ram_match = re.search(r'(\d+) ?(?:GB|MB)', ram)
        ram_data = ram_match.group(1) + " GB"
        
    def get_shipping_type(ad):
        try:
            shipping_data = ad.find_element(By.CSS_SELECTOR, '.nbd1Bd span')
            shipping_type = shipping_data.text
            if not shipping_type:
                shipping_type = 'NA'
        except NoSuchElementException:
            shipping_type = 'NA'
        return shipping_type

    for ad in ads:
        shipping_type = get_shipping_type(ad)

    # Append the data to the respective lists
    model_names.append(model_name)
    ram_datas.append(ram_data)
    prices.append(price)
    ecom_sites.append(ecom_site)
    shipping_types.append(shipping_type)
    image_urls.append(image_url)
    product_links.append(product_link)
    

# Create a DataFrame to store the scraped data
data = {
    'Model Name': model_names,
    'RAM': ram_datas,
    'Price': prices,
    'Ecom Site': ecom_sites,
    'Shipping Type': shipping_types,
    'Image URL': image_urls,
    'Product Link': product_links
}

df = pd.DataFrame(data)
df
```

```python
# Saving the data to an excel file
current_datetime = datetime.now()
timestamp = current_datetime.strftime("%Y-%m-%d_%H-%M-%S")
filename = f"ad_scraped_data_{timestamp}.xlsx"

df.to_excel(filename, index=False, engine='openpyxl')
```
